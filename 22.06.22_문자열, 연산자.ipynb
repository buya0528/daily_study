### 문자열 포매팅

```python
day = 50
s= '오늘은 7월 %d일 입니다.'%day
s
# %d 는 10진수로 정수 입력가능
```

* %s 문자열 
* %c 문자1개
* %d 정수
* %f 실수
* %% 문자 그 자체

##### 2개 이상의 값을 포매팅

```python
month = 7
day = 7
s= '오늘은 %d월 %d일 입니다'%(month,day)
s
```

##### format 함수를 사용하여 포매팅

```python
month = 7
day = 7
s= '오늘은 {}월 {}일 입니다'.format(month,day)
s
```

##### f-string 포매팅

```python
month = 7
day = 7
s= f'오늘은 {month}월 {day}일 입니다'
s
```

```python
x=100
y=200
sum2=x+y
print(f"{x}와 {y}의 합은 {sum2}입니다")
```

# 자주 사용하는 문자열 함수

* count함수 : 문자열에 포함된 문자 개수 세기

```python
s='오늘 점심 메뉴는 탕수육 입니다. 저는 탕수육을 매우 좋아해서 탕수육만 먹을겁니다.'
```

```python
s.count('탕수육')
```

* strip 함수 : 양쪽 공백 지우기

```python
s = '      오늘 점심 메뉴는 탕수육 입니다   '
s
```

```python
s.strip()
```

```python
a='가나다 abc 가나다'
```

```python
a.strip('가나다')
```

* replace 함수 : 특정 문자를 원하는 문자로 변경하기

```python
s='오늘 점심 메뉴는 탕수육 입니다. 저는 탕수육을 매우 좋아해서 탕수육만 먹을겁니다.'
```

```python
s.replace('탕수육','마라탕')
```

* split 함수 : 문자열을 특정 문자 기준으로 나눠주기

```python
s.split(".")
```
